## Analysis of the popular app categories:

For the analysis of the popular app categories firts we display basic information of the dataset with the few rows of the dataset

```python
print(df.info())
```

```python
print(df.head())
```

for the analysis of the popular app categories firts we calculate the average number of ratings per genre

```python
avg_ratings_per_genre = df.groupby('prime_genre')['rating_count_tot'].mean()
```

Then we Sort the genres by average number of ratings in descending order

```python
popular_genres = avg_ratings_per_genre.sort_values(ascending=False)
```

```python
print(popular_genres.head())
```

We created a bar plot to visualize the most popular app categories based on the average ratings

```python
import matplotlib.pyplot as plt

# Create a bar plot
plt.figure(figsize=(10, 6))
popular_genres.head(10).plot(kind='bar', color='blue')
plt.title('Top 10 Popular App Categories based on Average Ratings')
plt.xlabel('App Category')
plt.ylabel('Average Ratings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## In-App Purchases and Revenue:

```python
print(df.columns)
```

Firts we Create a new column indicating whether an app has in-app purchases
Second we Count the number of apps with in-app purchases per genre
third we Display the number of apps with in-app purchases per genre

```python

df['has_in_app_purchases'] = df['price'].apply(lambda x: 'yes' if x != 0 else 'no')

apps_with_in_app = df.groupby('prime_genre')['has_in_app_purchases'].apply(lambda x: (x == 'yes').sum())

print(apps_with_in_app)
```

Calculate the potential revenue for apps with in-app purchases, Calculate the total potential revenue per genre and the sort by the genres

```python
df['num_in_app_purchases'] = df['price'].apply(lambda x: len(str(x).split(',')) if x > 0 else 0)
df['potential_revenue'] = df['price'] * df['num_in_app_purchases']

total_potential_revenue = df.groupby('prime_genre')['potential_revenue'].sum()

sorted_genres = total_potential_revenue.sort_values(ascending=False)

```

to view the data we created a bar plot 

```python

plt.figure(figsize=(10, 6))
sorted_genres.head(10).plot(kind='bar', color='orange')
plt.title('Top 10 Potential Revenue by App Category')
plt.xlabel('App Category')
plt.ylabel('Potential Revenue ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Correlation Analysis:
this wild help understand the relationships between different numeric variables in a dataset

firts we calculate the correlation matrix and then displayed 

```python
correlation_matrix = df.corr()

print(correlation_matrix)
```

Import seaborn and metplotlib.pyplot for visualization, then put the figure and axes and created the headmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```
